```python
from ufit import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
for i in range(90117, 90135+1):
    if i == 88884:
        continue
    _ = plt.figure()
    df = pd.read_csv(f"H_along_ab/1200mK/0{i}.txt", names=["QH", "QK", "QL", "CNTS", "1", "2", "3", "4", "5", "6", "T", "H", "7"], sep=",", dtype=np.float64)
    peaks, _ = find_peaks(df["CNTS"], height=70, distance=30)
    plt.plot(df["QL"], df["CNTS"], "bo")
    for point in peaks:
        plt.plot(df["QL"][point], df["CNTS"][point], "ro")
    plt.title(f"{df["H"][0]}")

plt.show()
```
