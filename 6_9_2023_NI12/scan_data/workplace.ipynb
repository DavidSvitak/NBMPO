```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.backends.backend_pdf import PdfPages
from ufit.lab import *
from ufit.plotting import mapping
%matplotlib inline
```

```python
set_datatemplate('070%03d')
#set_dataformat('ill')
data = read_numors('76-93', 0)
set_cmap("jet")
mapping("QH", "QL", data, mode=0, title="", label="Intensity (arb. unit)")
plt.xlabel("(Qh,Qh,0) (r.l.u.)")
plt.ylabel("(0,0,Ql) (r.l.u.)")
plt.ylim((0.05,0.3))
plt.xlim((0.2, 0.9))
plt.savefig("Elastic_scattering_55mK.pdf")

show()

```
