```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.backends.backend_pdf import PdfPages
from ufit.lab import *
from ufit.plotting import mapping
%matplotlib inline
```

```python
file = open(f'070116')
lines = file.readlines()

for _ in range(6):
    del lines[51]

with open(f'new_070116', 'w') as file:
    for line in lines:
        #print(line)
        file.write(line)

```

```python

for i in range(136, 146):
    if i < 100:
        file = open(f'0700{i}')
        lines = file.readlines()

        for _ in range(9):
            del lines[51]

        with open(f'new_0700{i}', 'w') as file:
            for line in lines:
                file.write(line)

    elif i >= 100:
        file = open(f'070{i}')
        lines = file.readlines()

        for _ in range(6):
            del lines[51]

        
        with open(f'new_070{i}', 'w') as file:
            for line in lines:
                file.write(line)




```

```python
set_datatemplate('new_070%03d')
data = read_numors('97-100, 102-106', 0)
data.name = 'Energy scans (h, h, 0.18)'

mapping("QH", "EN", data, mode=0, title=f'{data.name}', )  # log=True)
plt.savefig("Inelastic_scattering_at_(h,h,0.18).pdf")
show()
```

```python
set_datatemplate('new_070%03d')
data = read_numors('97, 136-137,138+139,140-145', 0)
data.name = 'Energy scans (1/3,1/3,l)'

mapping("QL", "EN", data, mode=0, title=f'{data.name}', )#log=True)
plt.savefig("Inelastic_scattering_in_c_direction.pdf")
show()
```
